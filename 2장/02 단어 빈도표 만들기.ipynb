## Type-Token Ratio (TTR)
* TTR = Types / Tokens
* TTR이 낮을수록 반복이 많거나 용어 사용이 단조로움
* 텍스트 내 어휘의 다양성을 알려주는 지표로 사용됨

````python
# Type-Token Ratio

types = len(set(sent))
tokens = len(sent)
TTR = types/tokens
print(TTR)
````

## 예제2

````python
import nltk
nltk.download('brown')
````

````python
from nltk.corpus import brown
````

````python
categories = brown.categories()
````

````python
for category in categories:
    # 해당 카테고리의 단어들을 가져옵니다.
    words = brown.words(categories=category)
    # 단어의 총 수
    tokens = len(words)
    # 고유한 단어의 수
    types = len(set(words))
    # 어휘 다양성 점수(TTR의 역수) 계산
    diversity_score = tokens / types if types != 0 else 0
    # string formatting
    table = "{:<20}| {:<100}| {:<10}"
    # print(f"Category: {category}, Diversity Score: {diversity_score}")
    print(table.format(category, repr(words), "{:.2f}".format(diversity_score)))
````

## 함수(function)
* 특정 작업을 순차적으로 수행하는 명령행 모음집
* 입력값(input)을 인수(arguments)로 받아서 작업을 수행한 후 결과값(output)을 반환(return) 함

````python
# 함수의 정의(인수1=분자, 인수2=전체빈도(분모))
def percentage(bunza, bunmo):
  return bunza / bunmo * 100
  #return str(count / total * 100) + '%'
````

````python
percentage(4, 10)
````

````python
def type_token_ratio(text):
  types = len(set(text)) # text의 유형 가짓수
  # print(types)
  tokens = len(text) # text의 토큰 수
  # print(tokens)
  TTR = types / tokens  # TTR 계산 공식
  return TTR
````

## 단어 빈도표 만들기
* FreqDist 함수

````python
from nltk.book import text3
````

````python
from nltk.probability import FreqDist
````

````python
FreqDist
````

* 에러 처리
  - `from nltk.book import *`
  -  `nltk.download()`
  - `import nltk`

````python
text3
````

````python
fdist = FreqDist(text3)
````

````python
# 가장 빈도수가 높은 Topn을 보이려면?
# help(fdist.most_common)
fdist.most_common?
````

````python
fdist.most_common(30) # him, his 성서는 남성 위주
````

````python
text3.count('God')
````

````python
fdist.plot(20)
````

````python
fdist.plot(10, cumulative=True, percents=True) # And까지 전체 토큰의 약 30% 가까이 차지
````

````python
fdist.plot(200, cumulative=True, percents=True)     # 3000개 중 10%도 안되는(200개)토큰들이 전체 텍스트 중 약 80% 차이
# 대부분의 분포는 이렇다
````

````python
# 가장 빈도수가 낮은 희소 단어(hapax legomena)를 보이려면?
# help(fdist.hapaxes)
fdist.hapaxes?
````
